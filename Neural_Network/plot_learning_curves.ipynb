```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_RNA = np.load("learning_curves/Learning_Curves_RNA_subset.npz")
data_ATAC = np.load("learning_curves/Learning_Curves_ATAC.npz")
```

```python
plt.figure(figsize=(20,7))
plt.rcParams.update({'font.size': 18})
plt.subplot(1,2,1)
plt.plot(np.arange(100), data_RNA['train_loss'], label="Training")
plt.plot(np.arange(100), data_RNA['test_loss'], label="Testing")
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim([0, 0.15])
plt.legend()

plt.subplot(1,2,2)
plt.plot(np.arange(100), data_ATAC['train_loss'], label="Training")
plt.plot(np.arange(100), data_ATAC['test_loss'], label="Testing")
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim([0, 0.15])
plt.legend()

plt.savefig("plots/Learning_Curves.png")
```
